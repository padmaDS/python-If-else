```python
def profit_flag(a) :
    if a >= 0 :
        return 'positive'
    else :
        return 'negative'
```

```python
def quantity_flag(a) :
    if  a < 10 :
        return 'low'
    elif a >10 and a <20 :
        return 'medium'
    else :
        return 'high'
```
